```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#replacing blanks with 0 as tenure is 0 and no total charges are recorded

```python
df["TotalCharges"]=df["TotalCharges"].replace(" ","0")
df["TotalCharges"]=df["TotalCharges"].astype("float")
```

```python
df["customerID"].duplicated().sum()
```

```python
def conv(value):
    if value==1:
        return 'Yes'
    else:
        return 'No'
df["SeniorCitizen"] = df['SeniorCitizen'].apply(conv)
```

```python
df.head(30)
```

#converted 0 and 1 values to yes/no to make it easier

```python
plt.figure(figsize=(3,4))
ax=sns.countplot(x='Churn',data=df,hue='Churn',palette='GnBu')
ax.bar_label(ax.containers[0])
plt.title('Count of Customers by Churn')
plt.show()
```

```python
plt.figure(figsize=(3,4))
color=sns.color_palette('GnBu',2)
gp=df.groupby('Churn').agg({'Churn':'count'})
plt.pie(gp['Churn'],colors=color,labels=gp.index,autopct="%1.2f%%")
plt.title('Percentage of Churned Customers',fontsize=10)
plt.show()
```

#from the given pie chart we conclude that 26.54% of customers are churned out.
#now let's explore the reason behind it

```python
plt.figure(figsize=(3,3))
plt.title('Churn by Gender')
sns.countplot(x='gender',data=df,hue='Churn',palette='GnBu')
plt.show()
```

```python
plt.figure(figsize=(3,4))
plt.title('Churn by Senior Citizens')
ax=sns.countplot(x='SeniorCitizen',data=df,hue='SeniorCitizen',palette='GnBu')
ax.bar_label(ax.containers[0])
plt.show()
```

```python
# Assuming df is your DataFrame with 'SeniorCitizen' and 'Churn' columns

# Calculate the percentage of each Churn category by SeniorCitizen
df_counts = df.groupby(['SeniorCitizen', 'Churn']).size().unstack(fill_value=0)
df_percentage = df_counts.div(df_counts.sum(axis=1), axis=0) * 100  # Convert to percentages

# Plotting the stacked bar chart
plt.figure(figsize=(2,2))
ax = df_percentage.plot(kind='bar',color=sns.color_palette('GnBu', 2) ,stacked=True, width=0.4)

# Set title and labels
plt.title('Churn by Senior Citizens(stacked bar chart)', fontsize=10)
plt.xlabel('Senior Citizen', fontsize=10)
plt.ylabel('Percentage (%)', fontsize=10)

# Adding percentage labels on top of each bar
for p in ax.patches:
    height = p.get_height()
    if height > 0:  # Add label only for non-zero height
        ax.annotate(f'{height:.1f}%', 
                    (p.get_x() + p.get_width() / 2., p.get_y() + height / 2.),
                    ha='center', va='center', fontsize=10, color='black', weight='normal')
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```

```python
# List of columns
columns = ['PhoneService', 'MultipleLines', 'InternetService', 
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 
           'TechSupport', 'StreamingTV', 'StreamingMovies']

# Number of subplots
num_columns = len(columns)
# Set the number of rows and columns for the subplots (e.g., 3 rows and 3 columns)
rows = (num_columns // 3) + (num_columns % 3 > 0)  # Calculate rows needed
cols = 3  # You can adjust the number of columns as per your preference

# Create subplots
fig, axes = plt.subplots(rows, cols, figsize=(12, 12))
axes = axes.flatten()  # Flatten the axes to make indexing easier

# Loop over each column and plot countplot
for i, col in enumerate(columns):
    sns.countplot(x=col, data=df, ax=axes[i], palette='GnBu',legend=False,hue=df['Churn'])
    axes[i].set_title(f'Countplot for {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')

# Remove any empty subplots (if any)
for j in range(i + 1, len(axes)):
    axes[j].axis('off')

# Adjust layout for better visualization
plt.tight_layout()
plt.show()
```

#the majority of customers who do not churn are likely to have services like PhoneService,
InternetService(particulalry DSL) and online security enabled. Other services churn rates are noticeably higher due to their unavailibility
